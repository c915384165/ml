## 约定 (Deep neural network notation)

4 layers NN with 3 hidden layers

Capital L means layers, here 

$L=4$ (#layers=number of layers) (#: number of __)

### 输入特征
$X=a^{[0]}$ $n^{[0]}=n_x=3$,

### 节点数
$n^{[l]}$ (#units(or nodes))

here:

$n^{[0]}=n_x=3$, $n^{[1]}=5$, $n^{[2]}=5$, $n^{[3]}=3$, $n^{[4]}=1=n^{[L]}$

## 激活函数
$a^{[l]}=g^{[l]}(Z^{[l]})$ 

$a^{[l]}$: activations in layer l

$X=a^{[0]}, n_x=n_0$

$L$:(#layers)

$n^{[l]}$: (#units)

$a^{[l]}$ activations in layer l

$a^{[l]}=g^{[l]}(Z^{[l]})$

$W^{[l]}$: weight for $Z^{[l]}$

$b^{[l]}$

# 4.2 深层网络中的前向传播 

$X=a^{[0]}$

X-> Layer 1

$Z^{[1]}=W^{[1]}a^{[0]}+b^{[1]}$

$a^{[1]}=g^{[1]}(Z^{[1]})$

Layer 1 -> Layer 2

$Z^{[2]}=W^{[2]}a^{[1]}+b^{[2]}$

$a^{[2]}=g^{[2]}(Z^{[2]})$

Layer l-1 -> Layer l

$Z^{[l]}=W^{[l]}a^{[l-1]}+b^{[l]}$

$a^{[l]}=g^{[l]}(Z^{[l]})$

$\hat{Y}=g(Z^{[L]})=A^{[L]}$

```python
# 添加中间层、输出层节点
for layer_index, layer in enumerate(all_nodes[1:]):
    color = "#5bc49f" if layer_index == len(all_nodes[1:]) - 1 else "#ff7c7c"
    print(f"layer_index:{layer_index}")
    for node_index, node in enumerate(layer):
        x = 1.5 * (layer_index + 1)
        y = - 0.3 * (len(all_nodes[layer_index + 1]) - len(all_nodes[0])) + 0.6 * node_index
        G.add_node(node, label="", shape="circle", style="bold", color=color, pos=f"{x},{y}!")
#         print(f"{x},{y}")
        print(f"node_index:{node_index}")
        print()
#         print(f"node:{node}")
        print(f"x = 1.5 * ({layer_index} + 1)")
        print(f"y = - 0.3 * ({len(all_nodes[layer_index + 1])} - {len(all_nodes[0])}) + 0.6 * {node_index}")
        print(f"x:{x},y:{y}")
    # 添加中间层、输出层文本
    text = f"Output_Layer({len(all_nodes[-1])})" if layer_index == len(all_nodes[1:]) - 1 else f"Layer_{layer_index + 1}({len(layer)})"
    G.add_node(f"Layer_{layer_index + 1})", shape="none", label=text,
               pos=f"{x},{y + 0.4}!")
#     print(pos)
```

```python
# 添加线
for layer_index in range(all_nodes.__len__() - 1):
    for start in all_nodes[layer_index]:
        for end in all_nodes[layer_index + 1]:
            G.add_edge(start, end)
```
