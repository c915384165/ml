## 2.2 Logistic Regression (Logistic 回归)

Given x, want $\hat{y}=P(y=1|x)$, $\hat{y}\in[0,1]$

$\hat{y}$ 表示预测值，相当于关于x的概率函数，当x满足条件时候y=1。

$x\in\mathbb{R}^{n_{x}}$

Parameters: $w\in\mathbb{R}^{n_{x}}$, $b\in\mathbb{R}$

Output: $\hat{y}=\sigma(w^{T}x+b)$, $\sigma(z)=\frac{1}{1+e^{-z}}$
